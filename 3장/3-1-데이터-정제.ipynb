# 정규식

```python
import re
```

```python
test = '12이3이abc이#@것이54지!!#@!'
```

```python
# test에 있는 숫들만 모드 찾겠다.
# \d+ 모든 숫자들
re.findall('\d', test)
```

```python
# 뒤에 +가 써있으면 위와 다르게 아래와 같이 붙어서 나온다.
re.findall('\d+', test)
```

```python
# 숫자
int("".join(re.findall('\d+', test)))
```

```python
# a-z a에서 z까지 모든 것들 A-Z 모든 대문자 0-9 모든 숫자
# -는 앞에서 뒤까지모든 것들
re.findall('[a-zA-Z0-9]+', test)
```

```python
# ^이게 아닌것이라는 뜻 !랑 똑같다고 보면 됩니다.
# 따라서 모든 숫자와 영어가 아닌것들
re.findall('[^a-zA-Z0-9]+', test)
```

```python
"".join(re.findall('[a-zA-Z0-9]', test))
```

```python
"".join(re.findall('[^a-zA-Z0-9]', test))
```

```python
#자음 모음과 글자
re.findall('[ㄱ-ㅣ가-힣]+', test)
```

```python
"".join(re.findall('[ㄱ-ㅣ가-힣]+', test))
```

## 정규식 + map 함수

```python
def redata(test):
    data = re.findall('[0-9]+', test)
    return "/".join(data)
```

```python
redata("2018-11-20")
```

```python
def redata(test):
    data = re.findall('[0-9]+', test)
    
    return data[0] + "년" + data[1] + "월" + data[2] + "일"

redata("2018-11-20")
```

```python
df["start-days"] = df["start-days"].map(redata)
df
```

```python
df["end-days"] = df["end-days"].map(redata)
df
```

```python
#한글 숫자 개행 띄어쓰기 외에 다른 문자가 있으면 아예 그 행를 제거하는 함수
def onlyko(test):
    data = re.findall('[ㄱ-ㅣ가-힣\n0-9 ]+', test)
    if "".join(data) == test:
        return True
    return False
```

```python
df[df["petition_overview"].map(onlyko)]
```
