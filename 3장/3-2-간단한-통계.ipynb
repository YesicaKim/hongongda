## 카테고리를 알아보자

```python
df["category"]
```

```python
set(df["category"])
```

```python
df[df["category"] == "일자리"]
```

```python
# columns 찾기
len(list(df[df["category"] == "일자리"]))
```

```python
# rows 찾기
len(df[df["category"] == "일자리"])
```

```python
category_dict = {}
for category in set(df["category"]):
    category_dict[category] = len(df[df["category"] == category])
category_dict
```

```python
#깔끔하게 비율보기
for k,v in category_dict.items():
    print(k, " : ", round(v / len(df) * 100, 2), "%")
```

```python
category_dict.values()
```

```python
max(category_dict.values())
```

```python
# 정렬할 값, 정렬을 할 기준점, 내림차순으로 만들어라
sorted(category_dict.items(), key = lambda keys : keys[1], reverse=True)

# keys = 키, 벨류 값을 지정한 이름
# keys[0] = 키값(카테고리), keys[1] = 벨류값(청원동의 수) 
```

```python
sorted(category_dict.items(), key=lambda keys : keys[1])
```
