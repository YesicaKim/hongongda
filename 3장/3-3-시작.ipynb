```python
import nltk
from konlpy.tag import Twitter; t = Twitter()
```

```python
import pandas as pd
df = pd.read_csv("test_dataset.csv")
df
```

```python
category = "일자리"
# 일자리 카테고리만 지정
df1= df[df["category"]==category]
```

```python
text = list(df1["petition_overview"]) # 우리가 일자리 카테고리만 나오게한 청원 내용 전체를 list로 변경
text = ' '.join(text) # list로 바뀐 모든 청원 내용을 1개의 문자열로 만들기
ko_con_text = text # 변수명 번경
print(len(text)) #길이 확인
```

```python
tokens_ko = t.nouns(ko_con_text) # 모든 청원 내용을 형태소 단위로 바꾸는 코드
tokens_ko
```

```python
# 필요없는 것들 제거하는 코드 stop_words = 제거할 형태소를 넣는 리스트
stop_words = ['각력', '직','제','저', '이','그','때','및','것이','합니다.','\\n',':','','거','수','그' ,"합","이","있는","있습니다.","것","및","하지만","때문에"]

tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words] 

#for each_word in tokens_ko:
#    if each_word not in stop_words:
#        tokens_ko.append(each_wordword)
```

```python
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name= category+' 청원') #형태소 단위로 나눈 리스트, 이름 지정
data = ko.vocab().most_common(500) #얼마나 각각의 형태소들이 등장했는지 체크, 상위 500개만 남기고 제거
tmp_data = dict(data) # 딕셔너리 형태로 바꾸는 코드
tmp_data
```

```python
def test(category):
    df1= df[df["category"]==category]
    text = list(df1["petition_overview"]) # 우리가 일자리 카테고리만 나오게한 청원 내용 전체를 list로 변경
    text = ' '.join(text) # list로 바뀐 모든 청원 내용을 1개의 문자열로 만들기
    ko_con_text = text # 변수명 번경
    print(len(text)) #길이 확인
    tokens_ko = t.nouns(ko_con_text) # 모든 청원 내용을 형태소 단위로 바꾸는 코드
    tokens_ko
    # 필요없는 것들 제거하는 코드 stop_words = 제거할 형태소를 넣는 리스트
    stop_words = ['각력', '직','제','저', '이','그','때','및','것이','합니다.','\\n',':','','거','수','그' ,"합","이","있는","있습니다.","것","및","하지만","때문에"]

    tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words] 

    #for each_word in tokens_ko:
    #    if each_word not in stop_words:
    #        tokens_ko.append(each_wordword)
    ko = nltk.Text(tokens_ko, name= category+' 청원') #형태소 단위로 나눈 리스트, 이름 지정
    data = ko.vocab().most_common(500) #얼마나 각각의 형태소들이 등장했는지 체크, 상위 500개만 남기고 제거
    tmp_data = dict(data) # 딕셔너리 형태로 바꾸는 코드
    return tmp_data
```
