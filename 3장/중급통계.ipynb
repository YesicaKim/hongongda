```python
import nltk
from konlpy.tag import Twitter; t = Twitter()
```

```python
df
```

```python
def test(category, top_n):
    if category == "All":
        test_df = df
    else:
        test_df = df[df["category"]==category]
    text = list(test_df["petition_overview"]) # 우리가 지정한 카테고리만 나오게한 청원 내용 전체를 list로 변경
    text = ' '.join(text) # list로 바뀐 모든 청원 내용을 1개의 문자열로 만들기
    keywords = t.nouns(text) # 모든 청원 내용에서 명사를 추출하는 코드

    # 필요없는 것들 제거하는 코드 del_keywords = 제거할 형태소를 넣는 리스트
    del_keywords = ['각력', '직','제','저', '이','그','때','및','것이','합니다.','\\n',':','','거','수','그' ,"합","이","있는","있습니다.","것","및","하지만","때문에"]
    keywords = [keyword for keyword in keywords if keyword not in del_keywords] 

    keywords = nltk.Text(keywords) # 자연어 처리를 위해서 정의
    keywords = keywords.vocab()  # 각각의 명사가 몇 개 등장했는지 체크
    keywords = keywords.most_common(top_n) #상위 top_n개 남기고 제거 + 정렬
    keywords = dict(keywords) # 딕셔너리 형태로 바꾸는 코드

    return keywords
```

```python
top_keyword = test("All", 100)
top_keyword
```

```python
set(df["category"])
```

```python
categorys = list(set(df["category"]))
```

```python
keywords = {}
for category in categorys:
    # 카테고리 별로 상위 2000 개 뽑기
    keywords[category] = test(category, 2000)
```

```python
top_keyword_100 = list(top_keyword.keys())
top_keywords = {}

for category in categorys:
    top_keyword_dict = {}
    for keyword in top_keyword_100:
        # 상위 100개의 키워드가 각각의 카테고리별 키워드에서 몇 개 씩 나왔는지 확인
        try:
            top_keyword_dict[keyword] = keywords[category][keyword]
        except:
            top_keyword_dict[keyword] = 0
    top_keywords[category] = top_keyword_dict
```

```python
top_keywords
```

```python
categorys_dict = {}
for category in categorys:
    # Pandas로 만들기 위해서 values 값만 즉 숫자(등장 횟수)만 가져오기.
    categorys_dict[category] = list(top_keywords[category].values())
```

```python
test_df = pd.DataFrame(categorys_dict)
test_df.index = list(top_keyword.keys())
test_df
```
